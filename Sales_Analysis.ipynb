```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='latin1')
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
countplot = sns.countplot(x = 'Gender', data = df)

for bars in countplot.containers:
    countplot.bar_label(bars)
```

```python
df
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gender)
```

From above graph we can conclude that most of the buyers arev female. Strength of female is greater than male

```python
countplot1 = sns.countplot(x = 'Age Group', hue = 'Gender', data = df)

for bars in countplot1.containers:
    countplot1.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)


    
```

From the above graph we can conclude that most of the buyers are in 26-35 age group

State

```python
df
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(27,10)})
sns.barplot(x = 'State', y = 'Amount', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False)
sns.set(rc={'figure.figsize':(27,10)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```
